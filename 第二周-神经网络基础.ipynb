python例题：假设img是一个（32,32,3）数组，具有3个颜色通道：红色、绿色和蓝色的32x32像素的图像。 如何将其重新转换为列向量？

```python
import numpy as np

img = np.random.rand(32, 32, 3)

print("img.shape:\n")
print(img)
print(img.shape)

x = img.reshape(-1, 1) # -1 是 NumPy 的一种“自动推理维度”的写法，会根据剩下的维度自动计算出具体数值。
y = img.reshape((32*32*3), 1)

print(x)
print(y)

print(x.shape) # (3072, 1)
print(y.shape) # (3072, 1)
```
